```python
#how much on an average active users spend on phone durinf day time
print('active users spend on phone during day time : ',data[data['churn']==0]['total day minutes'].mean())
#how much on an average churn users spend on phone durinf day time
print('churn users spend on phone during day time : ',data[data['churn']==1]['total day minutes'].mean())

#how much on an average active users spend on phone durinf evening time
print('active users spend on phone during night time : ',data[data['churn']==0]['total eve minutes'].mean())
#how much on an average churn users spend on phone durinf nigth time
print('churn users spend on phone during night time : ',data[data['churn']==1]['total eve minutes'].mean())

#how much on an average active users spend on phone durinf night time
print('active users spend on phone during night time : ',data[data['churn']==0]['total night minutes'].mean())
#how much on an average churn users spend on phone durinf nigth time
print('churn users spend on phone during night time : ',data[data['churn']==1]['total night minutes'].mean())
```

```python
# get the average values for churn user
data[data['churn']==1].mean(numeric_only=True)
```

```python
# get the average values for active user
data[data['churn']==0].mean(numeric_only=True)
```

```python
#grouping

columns_to_show = ['total day minutes','total eve minutes','total night minutes']
data.groupby(['churn'])[columns_to_show].agg(np.mean)
```

```python

# Group data
grouped = data.groupby('churn')['total_minutes'].mean()

# Plot with colors
grouped.plot(kind='bar',color=['skyblue', 'orange'])  # Active vs Churn colors)

plt.title('Average Total Minutes: Active vs Churn Users')
plt.xlabel('Churn (0 = Active, 1 = Churn)')
plt.ylabel('Average Total Minutes Used')

plt.show()
```
